```python
from gensim import corpora
from gensim.models import LdaModel
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import pandas as pd
import re

# Function to preprocess text
def preprocess_text(text):
    # Lowercasing
    text = text.lower()
    # Removing punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenization
    tokens = text.split()
    # Removing stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return tokens

# Load the dataset
df = pd.read_excel('exam_questions(trial).xlsx', sheet_name='Easy')

# Apply preprocessing to the questions
preprocessed_questions = df['Question'].apply(preprocess_text)

# Create a dictionary representation of the questions
dictionary = corpora.Dictionary(preprocessed_questions)

# Create a document-term matrix
corpus = [dictionary.doc2bow(text) for text in preprocessed_questions]

# Define the number of topics
num_topics = 5

# Train the LDA model
lda_model = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=num_topics,
    passes=10,  # Increase the number of passes
    iterations=50  # You can also adjust the number of iterations
)

# Initialize list to store main topics
main_topics = []

# Extract main topics from each question
for text in preprocessed_questions:
    bow_vector = dictionary.doc2bow(text)
    topic_distribution = lda_model.get_document_topics(bow_vector)
    if topic_distribution:
        main_topic = max(topic_distribution, key=lambda x: x[1])[0]
        main_topics.append(main_topic)
    else:
        main_topics.append(None)  # Handle the case where no topic distribution is available

# Add main topics to the dataframe
df['Main Topic'] = main_topics

# Display the dataframe with main topics
print(df.head())
```
